```python
#Solving the multicollinearity between family size and number of children with vif

from statsmodels.stats.outliers_influence import variance_inflation_factor


def compute_vif(considered_features):
    
    X = df[considered_features]
    X['intercept'] = 1
    
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
```

```python
considered_features = ["YEARS_EMPLOYED","AGE","CNT_FAM_MEMBERS","FLAG_EMAIL","FLAG_WORK_PHONE","CNT_CHILDREN","AMT_INCOME_TOTAL","FLAG_PHONE","CODE_GENDER","FLAG_OWN_CAR","FLAG_OWN_REALTY","NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS","NAME_HOUSING_TYPE"]
compute_vif(considered_features).sort_values('VIF', ascending=False)
```

# Train-Test Split

```python
y=df_merged["Reject_costumer"].values
x=df_merged.drop("Reject_costumer",axis=1).values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import f1_score
```

# Classifier Algorithms
Will try resampling methods because the data is highly imbalanced.

1-DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_model = DecisionTreeClassifier(class_weight="balanced",max_depth=500,min_samples_split=22)
```

```python
DT_model.fit(x_train, y_train)
```

```python
y_pred = DT_model.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
y_pred_proba = DT_model.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```

# Conclusion
Randomforest with undersampling gives the best result to find the minority data. With recall value of 0.67 for finding the rejected costumers it is the highest one that in this project. If the bank's aim is to find who to reject this model can be used. However if the cost of incorrectly rejecting the costumer is too high other methods can be used. As an example Decision Tree Classifier without resampling has higher accuracy rates. Because of the imbalanced nature of the data it supports to accept more costumer's credit card applications. With this model there will be more costumers getting accepted for credit card while with randomforest with undersampling there will be more correctly rejected costumers. In the implementation stage these results should be considered.
